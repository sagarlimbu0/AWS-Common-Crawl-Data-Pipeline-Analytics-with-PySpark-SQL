```python
from pyspark.sql import functions as F
from pyspark import SparkConf
from pyspark import SparkContext
from pyspark.sql import SparkSession

from pyspark.sql import types
from pyspark.sql.types import StringType

import pandas as pd
import numpy as np
```

#### Configuration

```python
config = SparkConf()\
    .setAppName('AWS_Crawl')\
    .setMaster('local[*]')
```

#### Spark Session

```python
spark = SparkSession \
    .builder \
    .config(conf = config)\
    .getOrCreate()
```

#### Read file from web crawl

```python
# Path to data source
path= "../../data_aws/cdx-00009.gz"
```

```python
# df_web = spark.read\
#     .option('header', True)\
#     .csv("../data_sources/AWS_web_crawl/cdx-00000")

df_web = spark.read\
    .csv(path)
```

```python
df_web.show(2)
```

#### Get `unique` values from selected column 

```python
# df_web.select('_c0').distinct().collect()
```

### Rename `appropriate` column names

```python
## shcema
df_web = df_web\
    .withColumn("mime_type", F.col("_c2").cast(StringType())).drop("_c2")\
    .withColumn("content_mime_detected", F.col("_c3").cast(StringType())).drop("_c3")\
    .withColumn("fetch_status", F.col("_c4").cast(StringType())).drop("_c4")\
    .withColumn("content_digest", F.col("_c5").cast(StringType())).drop("_c5")\
    .withColumn("warc_record_length", F.col("_c6").cast(StringType())).drop("_c6")\
    .withColumn("warc_record_offset", F.col("_c7").cast(StringType())).drop("_c7")\
    .withColumn("warc_filename", F.col("_c8").cast(StringType())).drop("_c8")\
    .withColumn("content_charset", F.col("_c9").cast(StringType())).drop("_c9")\
    .withColumn("content_languages", F.col("_c10").cast(StringType())).drop("_c10")


    # .withColumn("_c0", F.col("_c0").cast(StringType())).drop("_c0")\
    # .withColumn("_c1", F.col("_c1").cast(StringType())).drop("_c1")\
    # .withColumn("_c11", F.col("_c11").cast(StringType()))\
    # .withColumn("_c12", F.col("_c12").cast(StringType()))
```
